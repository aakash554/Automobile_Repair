```python
df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("s3://databricks-project-jman/Auto repair dataset/customer_survey.csv")

catalog_name = "auto_repair_bronze" 
schema_name = "default"  
table_name = "customer_survey"

df.write.mode("overwrite").saveAsTable(f"{catalog_name}.{schema_name}.{table_name}")

print(f"Table created: {catalog_name}.{schema_name}.{table_name}")
```
