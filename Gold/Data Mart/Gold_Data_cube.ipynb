```python
%sql
CREATE OR REPLACE TABLE automobile_repair.gold.data_cube_comprehensive AS
SELECT 
  -- All fields from consolidated fact table
  f.*,
  
  -- Store Dimension
  ds.store_key,
  ds.store_name,
  
  -- Invoice Date Dimension
  dd_inv.date_key as invoice_date_key,
  dd_inv.quarter as invoice_quarter,
  dd_inv.day as invoice_day,
  dd_inv.day_name as invoice_day_name,
  dd_inv.month_name as invoice_month_name,
  dd_inv.year_quarter as invoice_year_quarter,
  
  -- Order Date Dimension
  dd_ord.date_key as order_date_key,
  dd_ord.quarter as order_quarter,
  dd_ord.day_name as order_day_name
  
FROM automobile_repair.gold.fact_orders_consolidated f

-- Join Store Dimension
LEFT JOIN automobile_repair.gold.dim_store ds 
  ON f.store_id = ds.store_id

-- Join Date Dimension for Invoice Date
LEFT JOIN automobile_repair.gold.dim_date dd_inv 
  ON f.invoice_date = dd_inv.full_date

-- Join Date Dimension for Order Date
LEFT JOIN automobile_repair.gold.dim_date dd_ord 
  ON f.order_date = dd_ord.full_date
```
